```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# local에서 한글 폰트를 사용하기 위한 설정
fe = fm.FontEntry(
    fname=r'/Users/gimsanghae/Library/Fonts/NotoSansKR-Regular.otf',  # otf 파일이 저장되어 있는 경로
    name='NotoSansKR')  # 이 폰트의 원하는 이름 설정
fe2 = fm.FontEntry(
    fname=r'/Users/gimsanghae/Library/Fonts/NotoSansKR-Bold.otf',  # otf 파일이 저장되어 있는 경로
    name='NotoSansKRBold')  # 이 폰트의 원하는 이름 설정

fm.fontManager.ttflist.insert(0, fe)  # Matplotlib에 폰트 추가
fm.fontManager.ttflist.insert(1, fe2)  # Matplotlib에 폰트 추가
plt.rcParams.update({'font.size': 15, 'font.family': 'NotoSansKR'})  # 폰트 설정
plt.rcParams.update({'font.size': 15, 'font.family': 'NotoSansKRBold'})  # 폰트 설정
```

```python
labels = ['A', 'B', 'C', 'D', 'E'] # label list
x = np.array([50, 20, 15, 10, 5]) # data numpy array
plt.pie(x, labels=labels) # pie chart(data, label)
plt.show() # show the pie chart
```

```python
plt.rc('font', family='NotoSansKR')  # run command를 의미, font 설정
x2 = [i for i in range(0,60,10)]
labels2 = ['A', 'B', 'C', 'D', 'E', 'F']
plt.pie(x2,labels=labels2, shadow=False,autopct='%0.1f %%', startangle=80)
plt.title('Testing')
plt.show()
```

```python
plt.rc('font', family='NotoSansKRBold')  # run command를 의미, font 설정
plt.pie(x, labels=labels, shadow=True, startangle=90, autopct='%0.1f %%')
# autopct: 퍼센트 표시 형식, startangle: 시작 각도, shadow: 그림자 표시
# %0.1 %%f: 소수점 첫째자리까지 표시
# %% : 실제 '%' 문자를 출력하려면 '%%'를 사용합니다. 첫 번째 '%'는 문자열 포맷팅을 나타내고, 두 번째 '%'는 실제 '%' 문자를 나타냅니다.
plt.title('시장점유율')
plt.show()
```

```python
d1 = np.random.randint(low=1, high=11, size=50) # 1~10 사이의 정수 50개 from numpy
d2 = d1 + np.random.randint(1, 5, size=d1.size)
data = np.column_stack((d1, d2,d2-d1))
data
```

# **Line V.S. Scatter**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read csv movies_by_year
movies_by_year = pd.read_csv('https://raw.githubusercontent.com/data-8/textbook/main/assets/data/movies_by_year.csv')
movies_by_year # Total Gross : 총 수익, Number of Movies : 영화 수
```

```python
path_data = "https://raw.githubusercontent.com/mlee-pnu/IDS/main/FDS07/"
actors = pd.read_csv(path_data + 'actors.csv')
actors # Total Gross : 총 수익, Number of Movies : 출연 영화 수, Average per Movie : 영화당 평균 수익, #1 Movie : 가장 많이 벌어들인 영화, Gross : #1 Movie의 개런티
```

```python
# plotting a line graph
print("Line graph: ")
plt.plot(movies_by_year["Year"], movies_by_year["Number of Movies"])
# x축은 Year, y축은 Number of Movies
plt.show()

# plotting a scatter plot
print("Scatter Plot:  ")
plt.scatter(actors["Number of Movies"], actors["Total Gross"])
# x축은 Number of Movies, y축은 Total Gross
plt.show()
```

```python
actors.plot.scatter(x='Number of Movies', y='Total Gross')
# x축은 출연 영화 수, y축은 총 수입을 의미한다.
```

```python
# plot style
plt.plot(movies_by_year["Year"], movies_by_year["Number of Movies"], linewidth=3.0, label='# of Moveies')
# x축은 연도, y축은 해당 연도의 상영 영화 수, 선굵기 3.0, label은 # of Movies
plt.plot(movies_by_year["Year"], movies_by_year["Total Gross"], linewidth=2.0, label='Total Gross')
# x축은 연도, y축은 연도 당연도 총 수입, 선 굵기 2.0, label은 Total Gross
plt.title("Eaxample: Meaningless Comparsion")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.legend()
plt.axis([1979, 2016, 0, 12000]) # year : 1979 ~ 2016, Number of Movies : 0 ~ 12000
plt.grid(axis='y', color='purple', linestyle='--', linewidth=1)
# Grid 설정
plt.show()
```

```python
# oo style
# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
# Object Orient style?

# layout constrained 서로 겹치지 않게, constrained 값만 존재한다.
fig, ax = plt.subplots(figsize=(5, 3), layout='constrained')
ax.plot(movies_by_year["Year"], movies_by_year["Number of Movies"], linewidth=3.0, label='# of Moveies')
ax.plot(movies_by_year["Year"], movies_by_year["Total Gross"], linewidth=2.0, label='Total Gross')
ax.set_title("Eaxample: Meaningless Comparsion")
ax.set_xlabel("Year")
ax.set_ylabel("Number of Movies")
ax.legend()
ax.set_xlim([1979, 2016])
ax.set_ylim([0, 12000])
# ax.grid(axis = 'y', color = 'purple', linestyle = '--', linewidth = 1)
ax.grid(True, linestyle='--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$", # \mu 는 평균을 의미한다.
)

print(type(species))
weight_counts = {
    "Below": np.array([70, 31, 58]), # 평균 아래
    "Above": np.array([82, 37, 66]), # 평균 위
}
print(type(weight_counts))
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count  # 각각의 value들 복잡하기 때문에 거의 쓰지 않는다.

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

# example data
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(437)

num_bins = 30

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=True, ec='black')

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu)) ** 2))
ax.plot(bins, y, '--')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

mu_x = 200
sigma_x = 25
x = np.random.normal(mu_x, sigma_x, size=100)

mu_w = 200
sigma_w = 10
w = np.random.normal(mu_w, sigma_w, size=100)

fig, axs = plt.subplots(nrows=2, ncols=2)

axs[0, 0].hist(x, 20, density=True, histtype='stepfilled', facecolor='g',
               alpha=0.75)
axs[0, 0].set_title('stepfilled')

axs[0, 1].hist(x, 20, density=True, histtype='step', facecolor='g',
               alpha=0.75)
axs[0, 1].set_title('step')

axs[1, 0].hist(x, density=True, histtype='barstacked', rwidth=0.8)
axs[1, 0].hist(w, density=True, histtype='barstacked', rwidth=0.8)
axs[1, 0].set_title('barstacked')

# Create a histogram by providing the bin edges (unequally spaced).
bins = [100, 150, 180, 195, 205, 220, 250, 300]
axs[1, 1].hist(x, bins, density=True, histtype='bar', rwidth=0.8)
axs[1, 1].set_title('bar, unequal bins')

fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# loading dataset
iris = sns.load_dataset('iris')
iris.shape # petal 꽃잎 # sepal 꽃받침
```

```python
iris.head()
```

```python
iris.groupby('species').count() # 종으로 묶어서 종의 개체 수를 센다
```

```python
c1 = iris[iris['species'] == 'setosa']
c2 = iris[iris['species'] == 'versicolor']
c3 = iris[iris['species'] == 'virginica']
```

```python
plt.boxplot((c1['petal_length'], c2['petal_length'], 
             c3['petal_length']))
plt.xticks([1, 2, 3], ['setosa', 'versicolor', 'virginica'])
plt.grid(color='gray',linestyle='--', linewidth=0.5)
plt.show()
```

```python
fig, ax = plt.subplots()
ax = sns.boxplot(data=iris, x='species', y='petal_length')
plt.show()
```
